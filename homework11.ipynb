```python
import os
import cv2
from time import time
import numpy as np
from sklearn.utils import shuffle
import tensorflow as tf
from tensorflow.keras import Model
from tensorflow.keras.layers import Dense, Input
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 6]
```

```python
# Step 1
folder = 'C:/r_d/subset_homework'
```

```python
# Traffic sign class 0
fnames_0 = os.listdir(os.path.join(folder, 'class_id_0'))
images_0 = [cv2.imread(os.path.join(folder, 'class_id_0', f), cv2.IMREAD_UNCHANGED) for f in fnames_0]
labels_0 = [0] * len(images_0)
# Showing 15 random samples
for cnt, idx in enumerate(np.random.randint(0, len(images_0), 15)):
    plt.subplot(3,5,cnt+1)
    plt.imshow(images_0[idx], cmap='gray', vmin=0, vmax=255)
    plt.title('Index : ' + str(idx)), plt.axis(False)
```

```python
# Traffic sign class 1
fnames_1 = os.listdir(os.path.join(folder, 'class_id_1'))
images_1 = [cv2.imread(os.path.join(folder, 'class_id_1', f), cv2.IMREAD_UNCHANGED) for f in fnames_1]
labels_1 = [1] * len(images_1)
# Showing 15 random samples
for cnt, idx in enumerate(np.random.randint(0, len(images_1), 15)):
    plt.subplot(3,5,cnt+1)
    plt.imshow(images_1[idx], cmap='gray', vmin=0, vmax=255)
    plt.title('Index : ' + str(idx)), plt.axis(False)
```

```python
images = images_0 + images_1
labels = labels_0 + labels_1
images, labels = shuffle(images, labels)
images = np.array(images)
labels = np.array(labels)

pixels = np.array([image.flatten() for image in images])/255

print('Number of samples class_0 : ', len(images_0))
print('Number of samples class_1 : ', len(images_1))
```

```python
# Step 2
# Creating the model
inputs = Input(shape=(pixels.shape[1]))
outputs = Dense(1, activation="sigmoid")(inputs)
model = Model(inputs, outputs)
tf.keras.utils.plot_model(model, to_file="model.png", show_shapes=True, show_layer_names=True)
model.compile(optimizer ='adam', loss = 'mean_squared_error')

# Showing Model Data
model.summary()
```

```python
# Trainning the model
history = model.fit(pixels, labels, epochs=500, batch_size=16)
# Plot training history
h = history.history
epochs = range(len(h['loss']))
plt.plot(epochs, h['loss'], '.-'), plt.grid(True)
plt.xlabel('epoch'), plt.ylabel('loss')
# Plot the weights
plt.plot(model.layers[1].weights[0].numpy(), '.-'), plt.grid(True)
print(model.layers[1].weights[1].numpy(), model.layers[1].bias.numpy())
```

```python
idx = 50
pred = model.predict(pixels[idx:idx+1, ...])
print(pred, labels[idx])
# Testing it on the entire dataset
predictions = model.predict(pixels).squeeze()
predictions = predictions > 0.5
correct = 0
for prediction, label in zip(predictions, labels):
    if prediction == label:
        correct = correct + 1

print('Accuracy : ', correct/len(labels))
for cnt, idx in enumerate(np.random.randint(0, len(images), 15)): 
    plt.subplot(3,5,cnt+1), plt.imshow(images[idx], cmap='gray', vmin=0, vmax=255)
    plt.title('Label: ' + str(labels[idx]) + ' | Prediction: ' + str(predictions[idx]))
    plt.axis(False)
```

```python
# The best accuracy I could achieve using 1 neuron is 0.9888143176733781
# Even when training for a longer time - it doesn't change
```

```python
# Step 3
# Creating the model
inputs = Input(shape=(pixels.shape[1]))
hidden = Dense(2, activation="sigmoid")(inputs)
outputs = Dense(1, activation="linear")(hidden) # model works much better if linear activation is used here
model = Model(inputs, outputs)
tf.keras.utils.plot_model(model, to_file="model.png", show_shapes=True, show_layer_names=True)
model.compile(optimizer ='adam', loss = 'mean_squared_error')

# Showing Model Data
model.summary()
```

```python
# Trainning the model
# It needs more time to train though
history = model.fit(pixels, labels, epochs=4000, batch_size=64)
# Plot training history
h = history.history
epochs = range(len(h['loss']))
plt.plot(epochs, h['loss'], '.-'), plt.grid(True)
plt.xlabel('epoch'), plt.ylabel('loss')
# Plot the weights
plt.plot(model.layers[1].weights[0].numpy(), '.-'), plt.grid(True)
print(model.layers[1].weights[1].numpy(), model.layers[1].bias.numpy())
```

```python
idx = 50
pred = model.predict(pixels[idx:idx+1, ...])
print(pred, labels[idx])
# Testing it on the entire dataset
predictions = model.predict(pixels).squeeze()
predictions = predictions > 0.5
correct = 0
for prediction, label in zip(predictions, labels):
    if prediction == label:
        correct = correct + 1

print('Accuracy : ', correct/len(labels))
for cnt, idx in enumerate(np.random.randint(0, len(images), 15)): 
    plt.subplot(3,5,cnt+1), plt.imshow(images[idx], cmap='gray', vmin=0, vmax=255)
    plt.title('Label: ' + str(labels[idx]) + ' | Prediction: ' + str(predictions[idx]))
    plt.axis(False)
```
